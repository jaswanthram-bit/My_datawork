```python
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv('C:/Users/jaswa/OneDrive/Desktop/733/data sets lab 1/DataAnalyst.csv')

# Assuming the column containing industries is named 'Industry'
# and the column containing salary estimates is named 'Salary Estimate'
# You should replace these with the actual column names in your dataset

# Filter out rows where the industry is not "-1"
df_filtered = df[df['Industry'] != "-1"]

# Define a function to extract salary values
def extract_salary(salary_string):
    try:
        return float(''.join(filter(str.isdigit, salary_string)))
    except:
        return None

# Apply the function to create 'Low Salary' and 'High Salary' columns
df_filtered['Low Salary'] = df_filtered['Salary Estimate'].apply(lambda x: extract_salary(x.split('-')[0]))
df_filtered['High Salary'] = df_filtered['Salary Estimate'].apply(lambda x: extract_salary(x.split('-')[1]))

# Get the top 5 industries (excluding "-1")
top_5_industries = df_filtered['Industry'].value_counts().nlargest(5).index.tolist()

# Filter DataFrame to include only top 5 industries
df_top_5_industries = df_filtered[df_filtered['Industry'].isin(top_5_industries)]

# Calculate the mean low and high salaries for each industry
mean_low_salary = df_top_5_industries.groupby('Industry')['Low Salary'].mean()
mean_high_salary = df_top_5_industries.groupby('Industry')['High Salary'].mean()

# Merge the mean salaries into a single DataFrame
mean_salaries = pd.DataFrame({'Mean Low Salary': mean_low_salary, 'Mean High Salary': mean_high_salary}).reset_index()

print("Mean Salaries by Industry:")
print(mean_salaries)
```

```python
mean_combined_salary = (mean_low_salary + mean_high_salary) / 2

# Merge the combined mean salary into the existing DataFrame
mean_salaries['Combined Mean Salary'] = mean_combined_salary

# Display the updated DataFrame
print("Mean Salaries by Industry:")
print(mean_salaries)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the DataFrame df_filtered from previous steps

# Get the top 5 industries based on some criteria (replace with your criteria)
top_5_industries = df_filtered['Industry'].value_counts().nlargest(5).index

# Filter DataFrame to include only top 5 industries
df_top_5_by_industry = df_filtered[df_filtered['Industry'].isin(top_5_industries)]

# Calculate the mean low salary estimates by industry and size
industry_size_low_salary = df_top_5_by_industry.groupby(['Industry', 'Size'])['Low Salary'].mean().reset_index()

# Calculate the mean high salary estimates by industry and size
industry_size_high_salary = df_top_5_by_industry.groupby(['Industry', 'Size'])['High Salary'].mean().reset_index()

# Create pivot tables for low and high salary estimates
pivot_table_low_salary = industry_size_low_salary.pivot(index='Industry', columns='Size', values='Low Salary')
pivot_table_high_salary = industry_size_high_salary.pivot(index='Industry', columns='Size', values='High Salary')

# Plot the data for low salary estimates
pivot_table_low_salary.plot(kind='bar', figsize=(12, 6))
plt.title('Mean Low Salary Estimates by Industry and Company Size (Top 5 Industries)')
plt.xlabel('Industry')
plt.ylabel('Salary Estimate (Low)')

# Show the plot for low salary estimates
plt.show()

# Plot the data for high salary estimates
pivot_table_high_salary.plot(kind='bar', figsize=(12, 6))
plt.title('Mean High Salary Estimates by Industry and Company Size (Top 5 Industries)')
plt.xlabel('Industry')
plt.ylabel('Salary Estimate (High)')

# Show the plot for high salary estimates
plt.show()




```
